```python
si = pint.UnitRegistry()
si.setup_matplotlib()

def rho(p, T, R):
    return p / (R * T)

R_specific = scipy.constants.R  *si.J / (si.mol * si.K) / (28.97 * si.g / si.mol)

print(rho_air:=rho(1000 * si.hPa, 300 * si.K, R_specific).to_base_units())

f"{rho_air.to(si.g / si.cm**3)=:.3g}"

pyplot.plot(T:=np.linspace(200, 400, 100)*si.K, rho(1000 * si.hPa, T, R_specific).to(si.g / si.cm**3))
pyplot.legend()
pyplot.xlabel(f'Tepmerature [{T.units:~}]')
show_plot()
```
